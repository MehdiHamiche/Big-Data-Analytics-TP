Consider the following optimization problem:

$$
\min_{x} x^2 -4x + 4
$$

1. Solve the problem analytically.  

2. Implement the objective function and return the image of 1.  

````python
# Answer 

def  f(x):
    return (x - 2)**2

print("f(1) = ", f(1))
````

2. Implement the derivative of the objective function.  

````python
# Answer 

def df(x):
    return(2*x - 4)
````

3. implement a method named ```gradient_descent``` that takes as parameters the initial point, the learning rate, the objective function, its gradient, and a number of iterations and performs the gradient descent method. The function should also return the precision, and the function evaluated in the minimum. 

````python
# answer

def gradient_descent(obj, grad, x0, alpha, M):
    xk = x0
    
    for m in range(M):
         xk = xk - alpha * grad(xk)
    
    precision = abs(grad(xk))
    
    return(xk, obj(xk), precision)
    
````

4. Test and evaluate the method to solve the problem above.

````python
# Answer 
gradient_descent(f, df, 6, 0.8, 100)
````

5. implement a method named ```gradient_descent_``` that takes as parameters the initial point, the learning rate, the objective function, its gradient, and a precision and performs the gradient descent method. The function should also return the number of iterations, and the function evaluated in the minimum. 

````python
# answer

def gradient_descent_(obj, grad, x0, alpha, eps):
    xk = x0
    m = 0
    precision = abs(grad(xk))
    while(precision > eps):
        xk = xk - alpha * grad(xk)
        precision = abs(grad(xk))
        m = m+1
    return(xk, obj(xk), m, precision)
````

6. Test and evaluate the method to solve the problem above.

````python
# Answer 
gradient_descent_(f, df, 6, 0.8, 0.01)
````

7. Visualize at each iteration the gradient descent method for different values for initialization, precision and learning rate.

````python
# Answer

import numpy as np
import matplotlib.pyplot as plt


def gradient_descent_(obj, grad, x0, alpha, eps):
    xk = x0
    m = 0
    precision = abs(grad(xk))
    xks = [x0]
    objs = [obj(x0)]
    while(precision > eps):
        xk = xk - alpha * grad(xk)
        precision = abs(grad(xk))
        m = m+1
        xks.append(xk)
        objs.append(obj(xk))
    return(xks, objs)



xks_values, obj_values = gradient_descent_(f, df, -6, 0.1, 0.01)

x = np.linspace(-6, 8, 1000)  
y = f(x)

# Create a figure and axis
fig, ax = plt.subplots()

ax.plot(x,y)
ax.scatter(xks_values,obj_values, marker = "x", color = "r")

# Set the title and axis labels
ax.set_title('Gradient descent visualization')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Add a grid
ax.grid()

# Display the plot
plt.show()
````
